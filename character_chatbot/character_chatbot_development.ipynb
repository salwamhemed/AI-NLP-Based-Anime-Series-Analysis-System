```python
import pandas as pd
import regex as re
from datasets import Dataset
```

```python
data_path = r"C:\Users\salwa\OneDrive\Desktop\Series_Analysis\data\hxh_episode_script.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
df = df.drop(df.columns[3], axis=1)  
df = df.drop(df.columns[4], axis=1)  
df = df.drop(df.columns[5], axis=1)  

df
```

```python
#Remove Actions from transcript
def remove_parenthesis(text):
    cleaned_text = re.sub(r'\(.*?\)', '', text)
    return cleaned_text
```

```python
df['line'] = df['line'].apply(remove_parenthesis)
df.head(50)
```

```python
df["number of words"] = df["line"].str.strip().str.split()
df["number of words"] = df["number of words"].apply(lambda x: len(x))
df.head()
```

```python
df["naruto response flag"] = 0
df.loc[(df["name"]=="Naruto") & (df["number of words"]>4),"naruto response flag"]= 1 
df.head()
```

```python
indexes_to_take = list(df[(df['naruto response flag']==1)&(df.index>0)].index)
```

```python
system_prompt = """  You are Naruto Uzumaki from the anime "Naruto." Respond as if you are him, embodying his energetic, determined, and optimistic personality. Your replies should reflect his unwavering spirit, love for friends, and dedication to becoming Hokage. Remember to maintain his signature enthusiasm and humor in your responses.
  """
```

```python
prompts = []
for i in indexes_to_take:
    prompt = system_prompt
    prompt += '\n'  
    prompt += df.iloc[i-1]['line'] + '\n'  
    prompt += df.iloc[i]['line'] 
    prompts.append(prompt)

```

```python
print(prompts[5])
```

```python
df = pd.DataFrame({"prompt":prompts})
df.head()
```

```python
dataset = Dataset.from_pandas(df)
```
